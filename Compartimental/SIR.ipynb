# Susceptible-Infected-Recovered (SIR) Model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
plt.rcParams.update({'font.size': 14})
```

```python
t = np.linspace(0,100,10000)
dt = 0.01
n_iter = 52

S_t = []
I_t = []
R_t = []
S0 = 0.98
I0 = 0.01
R0 = 0.01
beta = 0.4
gamma = 0.05

for i in range(len(t) + 1):
    if(i == 0):
        S = S0
        I = I0
        R = R0
    else:
        dS = -1 * beta * S * I * dt
        dI = (beta * S * I - gamma * I) * dt
        dR = gamma * I * dt
        
        S = S + dS
        print(S)
        I = I + dI
        R = R + dR

        S_t.append(S)
        I_t.append(I)
        R_t.append(R)

```

```python
plt.figure(figsize=(9,6))
#plt.title(f'Modelo SIR com $\\beta$ = {beta} e $\gamma$ = {gamma}')
plt.title(f'Modelo SIR')
plt.plot(t, S_t, label='S(t): Suscetíveis', linestyle='dashed')
plt.plot(t, I_t, label='I(t): Infectados', c='r')
plt.plot(t, R_t, label='R(t): Recuperados', linestyle='dotted', c='k')
plt.xlabel('Tempo (dias)')
plt.ylabel('Proporção de indivíduos')
plt.legend()
plt.savefig('sir-model.png')
```
